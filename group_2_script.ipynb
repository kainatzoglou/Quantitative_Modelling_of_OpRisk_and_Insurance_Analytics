```python
#Exercise1Ac
#Defining a function for the Poisson-Gamma to create the density and the cdf

def tweedie(n,lamda,alpha,beta):
    np.random.seed(seed=30)  
    
    rt=np.full(n,np.nan)
    N=np.random.poisson(lamda,n) #Poisson Distribution
    for i in range(n):
        rt[i]=np.random.gamma(N[i]*np.abs(alpha),beta) #Poisson-Gamma Model
    return(rt)
```

```python
#1Ac Example 1 with parameters of our choice for Poisson-Gamma density and cdf

#generated for n=10
n=10
lamda=3
alpha=2
beta=1
#print(tweedie(n,lamda,alpha,beta))

#plot of the density using the above parameters

s_tweedie=tweedie(n,lamda,alpha,beta)
sns.distplot(pd.DataFrame(s_tweedie),color="purple").set_title('Poisson-Gamma PDF for N=10,lamda=3,alpha=2,beta=1')
plt.xlabel('observations')
plt.ylabel('PDF')
plt.tight_layout()
plt.show()


#plot of the CDF (Distribution) using the above parameters

hist = np.histogram(s_tweedie)
hist_dist = scipy.stats.rv_histogram(hist)

a=hist_dist.cdf(np.min(s_tweedie))
print(a)
b=hist_dist.cdf(np.max(s_tweedie))
print(b)
X = np.linspace(0, 20)


plt.plot(X, hist_dist.cdf(X), label='CDF',color="purple")
plt.xlabel('observations')
plt.ylabel('CDF')
plt.title('Poisson-Gamma CDF for N=10,lamda=3,alpha=2,beta=1')
plt.show()


```

```python
#Exercise 1Bc
#Density function of the compound process given by the Poisson-Stable model

def lev(n,lamda,alpha,beta,gama,delta):
    np.random.seed(seed=40)
    
    rtl=np.full(n,np.nan)
    N=np.random.poisson(lamda,n)
    for i in range(n):
        #rtl[i]=levy.rvs(alpha,beta,loc=gamanew,scale=deltanew,size=1)
        dt=np.dot(N[i],N[i])
        gamanew=dt*gama
        deltanew=N[i]*delta+((dt*gama)-(N[i]*gama))
        rtl[i]=levy.rvs(loc=deltanew,scale=gamanew,size=1)
    return(rtl)
```

```python
#Exercise 1Bc Example 1
#1Bc Example 1 with parameters of our choice for Poisson-Levy density and cdf

n=10
lamda=2
alpha=0.5
beta=1
gama=1
delta=2
t=lev(n,lamda,alpha,beta,gama,delta)
#print(t)

s_levy=lev(n,lamda,alpha,beta,gama,delta)
sns.distplot(pd.DataFrame(s_levy),color="green").set_title('Poisson-Levy PDF for N=10,lambda=2,a=0.5,b=1,g=1,d=2')
plt.xlabel('observations')
plt.ylabel('PDF')
plt.tight_layout()
plt.show()

F2=np.sort(s_levy)
F3=np.array(range(10))/float(10)
plt.plot(F2,F3,color="green")
plt.title('Poisson-Levy CDF for N=10,lambda=2,a=0.5,b=1,g=1,d=2')
plt.xlabel('observations')
plt.ylabel('CDF')
```

```python
#Exercise 3Ab - Implementation 1
#The first order asymptotic approximation for the Poisson-Lognormal model tail function 
#of the Poisson-Lognormal LDA model

lamda=5 #Poisson parameter
s = 0.954 #Log-Normal parameter that produces the Log-Normal mu and sigma
fig, ax = plt.subplots(1, 1)
#generating the Log-Normal moments
mean, var, skew, kurt = lognorm.stats(s, moments='mvsk')
sr=np.sqrt(var) #computing the sigma
print(mean) #print the mean
print(sr) #print the sigma
#set the Log-Normal linspace
x = np.linspace(lognorm.ppf(0.01, s),lognorm.ppf(0.99, s), 100) 
survival=(1-lognorm.cdf(x, s)) #calculate the survival function
approximation=lamda*survival #mathematical type produced in 3Aa
ax.plot(x,approximation ,'r-', lw=5, alpha=0.6, label='lognorm cdf') #plot
plt.xlabel('observations')
plt.ylabel('Tail function')
plt.title('Poisson-Lognormal tail function for lambda=5,mu=1.57,sigma=1.92')
plt.show
```
